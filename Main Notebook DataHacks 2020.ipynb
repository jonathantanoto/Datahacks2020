```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Cleaning

```python
# importing data
population = pd.read_csv('SD1970_population.csv')
housing = pd.read_csv('SD1970_housing.csv')
```

```python
population
```

1.1. Dealing with null values, duplicate/ambiguous rows, datatype

```python
# dealing with null values, replace with NaN
population = population.replace("...", np.NaN)
population = population.replace('nan', np.NaN)
population.head()
```

```python
population.head()
```

```python
# removing commas in thousands
for col in population.columns:
    if population[col].dtype == object:
        population[col] = population[col].str.replace(",", "")
```

```python
population
```

```python
# changing column 0 to string
population.iloc[:, 0] = population.iloc[:, 0].astype(str)

# changing column 1 to int64
population.iloc[:, 1] = population.iloc[:, 1].astype('int64')

# changing column 2 to string
population.iloc[:, 2] = population.iloc[:, 2].astype(str)

# now for rest of columns
for col_idx in range(3, len(population.columns)):
    try:
        population.iloc[:,col_idx] = population.iloc[:,col_idx].astype('int64')
    except:
        population.iloc[:,col_idx] = population.iloc[:,col_idx].fillna(-1).astype('int64')
```

```python
population
```

Now we move on to doing the same thing (slightly different) with the housing dataset.

```python
housing
```

```python
# removing commas and dollar signs in thousands
for col in housing.columns:
    if col == "Tract name":
        continue
    if housing[col].dtype == object:
        housing[col] = housing[col].str.replace(",", "")
        housing[col] = housing[col].str.replace("$", "")
        housing[col] = housing[col].str.replace(".00", "")
```

```python
# dealing with null values, replace with NaN
housing = housing.replace(' ... ', np.NaN).replace('...', np.NaN).replace(' -   ', np.NaN).replace('', np.NaN).replace(' ', np.NaN).replace('nan', np.NaN)
housing
```

```python
housing
```

```python
# changing column 0 to string
housing.iloc[:, 0] = housing.iloc[:, 0].astype(str)

# changing column 1 to int64
housing.iloc[:, 1] = housing.iloc[:, 1].astype('int64')

# changing column 2 to string
housing.iloc[:, 2] = housing.iloc[:, 2].astype(str)

# now for rest of columns
for col_idx in range(3, len(housing.columns)):
    try:
        housing.iloc[:,col_idx] = housing.iloc[:,col_idx].astype('int64')
    except:
        housing.iloc[:,col_idx] = housing.iloc[:,col_idx].fillna(-1).astype('int64')
```

1.2 Concatenate population categories and simplify to female and male.

```python
# generating count for males and females per row
male_count = population.iloc[:, 4:26]
female_count = population.iloc[:, 26:48]

total_male_count_column = male_count.sum(axis = 1, skipna = True)
total_female_count_column = female_count.sum(axis = 1, skipna = True)
```

```python
# assigning them to the population data
population['Total Male Persons'] = total_male_count_column
population['Total Female Persons'] = total_female_count_column
```

## 2. Visualization

2.1 How big is each block/residential area? **What block is the second most common, following San Diego?**

```python
housing
```

This question can be answered by grouping the occurences of each place names.
Then we will look at the secont most occuring place following San Diego.

```python
occurence_table = housing.groupby('Place Name').count().iloc[:,[0]].rename(columns = {"Tract name": "Occurences"})
occurence_table
```

```python
occurence_table = occurence_table.sort_values(by=['Occurences'], ascending = False)
occurence_table
```

```python
bar = housing['Place Name'].value_counts().plot(kind='bar', x='Place Name', title="Occurences of each Place Name")

bar.set_xlabel("Place Name")
bar.set_ylabel("Frequency")
```

### Chula Vista comes in second, following San Diego for the most common place.
